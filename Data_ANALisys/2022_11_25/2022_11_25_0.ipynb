```python
import pandas as pd

DF = pd.read_csv("incidentiBis.csv")
print(len(DF[DF["età"]=="Desconegut"]))
temp = DF["età"][DF["età"]!="Desconegut"]
meanAge=int(temp.astype("int64").mean())

DF["età"].replace("Desconegut", meanAge, inplace=True)
DF["età"] = DF["età"].astype("int64")
DF["età"][DF["età"]>100]
DF["età"].replace(420,42,inplace=True)
print(DF["età"])
DF.boxplot("età")
```

```python
DF["genere"].unique()
#DF.plot.pie("genere")

DFtemp= pd.DataFrame(DF.groupby("genere").count())
DFtemp.iloc[:,0].plot.pie()
```
